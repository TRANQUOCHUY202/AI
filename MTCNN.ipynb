```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
name = '/content/drive/MyDrive/IMG_1218.jpg'
img = plt.imread(name)
detector = MTCNN()
face = detector.detect_faces(img)
plt.imshow(img)
ax = plt.gca()
print('so nguoi trong anh: ', len(face))
for face in face:
  x,y,width,height = face['box']
  rect = plt.Rectangle((x,y),width, height, fill = False , color='red')
  ax.add_patch(rect)
  for key,value in face ['keypoints'].items():
    dot = plt.Circle(value,radius=2,color= 'red')
    ax.add_patch(dot)


plt.show()
```

```python
import matplotlib.pyplot as plt
from mtcnn import MTCNN
def draw_faces(filename,result_list):
  data = plt.imread(filename)
  for i in range(len(result_list)):
    x1,y1, width, height = result_list[i]['box']
    x2,y2 = x1+width, y1+height
    plt.subplot(1,len(result_list),i+1)
    plt.axis('off')
    face=data[y1:y2,x1:x2]
    plt.imshow(face)
    #Lưu từng ảnh khuôn mặt khi cắt ra từ filename 
    plt.imsave('/content/drive/MyDrive/face{}.jpg'.format(i+1),face)
  plt.show()
filename = '/content/drive/MyDrive/IMG_1218.jpg'
img = plt.imread(filename)
detector = MTCNN()
faces = detector.detect_faces(img)
draw_faces(filename,faces)
```
